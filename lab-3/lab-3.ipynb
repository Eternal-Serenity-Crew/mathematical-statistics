# Задание 1. Сумма всех значений

```python
import csv

# Функция для подсчета суммы из CSV файла
def calculate_sum_from_csv(filename, column_name):
    total = 0
    with open(filename, 'r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            total += int(row[column_name])
    return total

# Подсчет сумм
sum_x = calculate_sum_from_csv('variant_7_sample_X.csv', 'x_i')
sum_y = calculate_sum_from_csv('variant_7_sample_Y.csv', 'y_j')

print(f"Сумма значений выборки X: {sum_x}")
print(f"Сумма значений выборки Y: {sum_y}")
print(f"Общая сумма: {sum_x + sum_y}")
```

# Задание 2. Среднее арифметическое

```python
def calculate_mean_from_csv(filename, column_name):
    total = 0
    count = 0

    with open(filename, 'r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            total += float(row[column_name])
            count += 1

    return total / count if count > 0 else 0

# Использование
mean_x = calculate_mean_from_csv('variant_7_sample_X.csv', 'x_i')
mean_y = calculate_mean_from_csv('variant_7_sample_Y.csv', 'y_j')

print(f"Среднее арифметическое выборки X: {mean_x:.4f}")
print(f"Среднее арифметическое выборки Y: {mean_y:.4f}")
```

# Задание 3. Сумма квадратов

```python
import math

def calculate_statistics(filename, column_name, operation='sum_of_squares'):
    # Чтение данных из файла
    data = []
    with open(filename, 'r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            data.append(float(row[column_name]))

    return sum(x ** 2 for x in data)


# Примеры использования
print(f"Сумма квадратов X: {calculate_statistics('variant_7_sample_X.csv', 'x_i')}")
print(f"Сумма квадратов Y: {calculate_statistics('variant_7_sample_Y.csv', 'y_j')}")
```

# Задание 4. Несмещенная выборочная дисперсия

```python
def unbiased_sample_variance(filename, column_name):
    # Чтение данных из файла
    data = []
    with open(filename, 'r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            data.append(float(row[column_name]))

    n = len(data)

    if n < 2:
        raise ValueError("Для вычисления дисперсии необходимо минимум 2 элемента")

    # Вычисление среднего арифметического
    mean = sum(data) / n

    # Вычисление суммы квадратов отклонений
    sum_squared_deviations = sum((x - mean) ** 2 for x in data)

    # Несмещённая дисперсия (деление на n-1)
    variance = sum_squared_deviations / (n - 1)

    return variance

# Использование
print("=== Несмещённая выборочная дисперсия ===")
var_x = unbiased_sample_variance('variant_7_sample_X.csv', 'x_i')
var_y = unbiased_sample_variance('variant_7_sample_Y.csv', 'y_j')

print(f"Дисперсия выборки X (s²_x) = {var_x:.4f}")
print(f"Дисперсия выборки Y (s²_y) = {var_y:.4f}")
```

# Задание 5. Квантиль распределение Стьюдента

```python
from scipy import stats

def student_quantile(alpha, df):
    """
    Вычисляет квантиль распределения Стьюдента.

    Параметры:
    alpha (float): уровень значимости (0 < alpha < 1)
    df (int): число степеней свободы

    Возвращает:
    float: квантиль распределения Стьюдента
    """
    # Для двустороннего критерия используем alpha/2
    # Для одностороннего - alpha
    quantile = stats.t.ppf(1 - alpha/2, df)
    return quantile

# Примеры использования
print("=== Квантили распределения Стьюдента ===")
print(f"α=0.05, df=110: t = {student_quantile(0.05, 110):.4f}")
```
